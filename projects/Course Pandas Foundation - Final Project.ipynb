```python
# TODO 05 - how much total sales, average sales, and standard deviation of sales your company make in 2017
sale_2017_agg = df[df['Order Date'].dt.year == 2017 ]['Sales'].agg(['sum','mean','std'])

sale_2017_agg   
```

```python
# TODO 06 - which Segment has the highest profit in 2018
# filter order date in 2018 .dt.year dt. work with data/time properties .year to extract year properties
# then groupb 'Segment' .sum to sum profit in that segment
# finally .idxmax() to find index that has maximum value
highest_profit_2018 = df[df['Order Date'].dt.year == 2018]\
    .groupby('Segment')['Profit'].sum()\
        .idxmax() 

highest_profit_2018
```

```python
# TODO 07 - which top 5 States have the least total sales between 15 April 2019 - 31 December 2019
top5_state_least_sales = df[(df['Order Date'] >= '2019-04-15') & (df['Order Date'] <= '2019-12-31')]\
    .groupby('State')['Sales'].sum().sort_values().head(5)
# in this cased I need top 5, so I sort first and use head()
top5_state_least_sales
```

```python
# TODO 08 - what is the proportion of total sales (%) in West + Central in 2019 e.g. 25% 
west_central_2019 = df[(df['Order Date'].dt.year == 2019) & (df['Region'].isin(['West','Central']))]['Sales'].sum()

total_sales = df['Sales'].sum()

proportion = (west_central_2019/total_sales)*100

print(f"{round(proportion,2)}%")
```

```python
import matplotlib.ticker as ticker # <--- NEW: You need this import

# find top 5 product sales
top5_products = df.groupby('Product Name')['Sales'].sum().nlargest(5)

# filter out top5 product
df_top5 = df[df['Product Name'].isin(top5_products.index)]

# group product name year and name
sales_trend = df_top5.groupby([df_top5['Order Date'].dt.year, 'Product Name'])['Sales'].sum().unstack()

# plot
ax = sales_trend.plot(kind = 'line')

# Force the x-axis to show only integer
ax.xaxis.set_major_locator(ticker.MaxNLocator(integer=True)) # <--- NEW: This sets the x-axis to integers

ax.legend(bbox_to_anchor=(0.5, -0.5), loc='lower center', title='Product Name') 
```
